## Problem 1:
### Problem Statement:Draw a pie plot based on number of wins for each team


```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data/Match.csv",encoding="ANSI")
```

```python
def orob(filename):
    df=pd.read_csv(filename)
    return df
orob("data/Match.csv")
```

```python
class win():
    def wins(p,filename):
        p.team=[]
        for col in df.values:
            p.team.append(col[10])
        p.uniqteam,p.countarr=np.unique(p.team,return_counts=True)
        p.plot()
    def plot(q):
        labels=q.uniqteam
        sizes=q.countarr
         
        plt.pie(sizes, labels=labels,autopct='%1.1f%%', shadow=True, startangle=0)

        plt.axis('equal')
        plt.show()
d=win()
d.wins("data/Match,csv")
        
        
```

```python
def orob(filename):
    df=pd.read_csv(filename)
    return df
orob("data/Match.csv")   
```

```python
class miwin():
    def wins(p,filename):
        p.year=[]
        for col in df.values:
            p.year.append(col[5])
        p.uniqyear=np.unique(p.year)
        p.arr=[]
        for i in p.uniqyear:
            p.count=0
            for col in df.values:
                if i==col[5] and 'Mumbai Indians'==col[9] and  'Mumbai Indians'==col[10]:
                    p.count+=1
            p.arr.append(p.count)
        p.plot()
    def plot(q):
        q.index = q.uniqyear
        plt.bar(q.index, q.arr)
        plt.xlabel('years', fontsize=5)
        plt.ylabel('matchs won', fontsize=5)
        plt.xticks(q.index,  fontsize=5, rotation=30)
        plt.title('No of matches won by mi ')
        plt.show()
t=miwin()
t.wins("data/Match.csv")
```
